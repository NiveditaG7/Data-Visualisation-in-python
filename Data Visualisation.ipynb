```python
# sacter plot is used to visualize the relationship btw 2 continous variables

plt.scatter(data=df,x='Attack',y='Defense')
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data=df,x='Attack',y='Defense',s=600,c='purple',marker='*')

#change axes range
plt.xlim(0,150)
plt.ylim(0,200)
#add title
plt.title('Relationship between Attack and Defense',size=20)
#add x and y labels
plt.xlabel('Attack Points',size=15)
plt.ylabel('Defense Points',size=15)
plt.grid(True)
#show plot
plt.savefig('DV.jpg')
plt.show()
```

```python
import numpy as np
plt.hist(x="Attack",data=df, bins=10,color='red')
plt.title('Attacks')
plt.show()
```

```python
plt.figure(figsize=(40,10))
plt.subplot(131)  #1 row 3 column 1st position
plt.hist(x="Attack",data=df, bins=10,color='red')
plt.title('Attacks')

plt.subplot(132) # 1 row 3 column 2nd poistion
plt.hist(x="Defense",data=df, bins=10,color='blue')
plt.title('Defense')
plt.xticks(np.arange(0,201,10),rotation=90) #array from 0 to 200 with a increment of 10
plt.yticks(np.arange(0,41,5))

plt.subplot(133) # 1 row 3 column 3rd poistion
plt.hist(x="Speed",data=df, bins=10,color='green')
plt.title('Speed')
plt.show()
```

```python
#line plot is useful to visualize a continous variable over a period of time
#example: stock market, sensex, ECG

x=[10,30,20,60]
y=[10,45,20,70]
plt.plot(x,y,c='purple')

plt.xlim(0,100)
plt.ylim(0,100)
plt.title('A Simple Line Plot')
plt.grid(True)
plt.savefig('test.jpg')
for a,b in zip(x,y):
    plt.text(a,b,(a,b))
plt.show()
```

```python
#pie chart are useful to visualize the % distribution of a single categorical variable

# slices will be plotted counter-clockwise:

labels =['SQL','Python','R','Tableau']
sizes=[75,156,40,30]
#sizes =[14.5 45.5 30,10]
explode =[0.2,0,0.1,0]
plt.pie(sizes,explode=explode, labels=labels, autopct="%1.2F%%",
       colors=["red","purple","pink","green"],
       shadow=True, startangle=90,counterclock=True)
                
plt.show()
```
